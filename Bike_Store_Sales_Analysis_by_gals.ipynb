# 🚴 📊 Order Quantity Histogram 📊 🚴

> 🚴 To visualize the distribution of order quantities, we created a histogram  🚴

```python
# figure.size untuk resizing the chart
plt.figure(figsize=(10,5))

# kde estimates the density function at a point x using neighboring observations
hist_OO = sns.histplot(x='Order_Quantity', data=sales, kde=True)

# customize the title
plt.xlabel('Order Quantity')
plt.ylabel('Frequency')
plt.title('Order Quantity Histogram', fontsize=16)

#show the chart
plt.show()
```

> This histogram reveals that most of our customers tend to make 🚴 **0 - 5 order quantities**, with 🛒 **over than 2000 frequencies** being the most common.  Most customers 👪 place small orders, with a smaller but significant proportion opting for higher quantities.

> Recognizing that a substantial portion of your customers prefers small orders provides an 💹 opportunity for targeted marketing

# 💸 Sales per Year 💸

>  💰 Taking a yearly perspective, we determined the Sales Performance over the Years 💰

```python
sales_per_year = sales['Year'].value_counts()
print(sales_per_year)
```

>  💰 To aiming the total sales performance per year, we turn into Pie Chart 💰

```python
colors = sns.color_palette('Set1')
sales_per_year.plot(kind='pie', colors=colors, figsize=(6,6),autopct='%1.1f%%', fontsize=10)

plt.title('Sales per year', fontsize=16)
plt.show()
```

> Our analysis unveiled that the 💹 **total sales figures have steadily increased from 2011 to 2016**. Sales performance had been a **generally consistent upward trajectory** ⭐, with varying growth rates.

>🔎 Analyzing the factors contributing to the periods of high growth and the cyclical pattern can help the business make informed decisions and strategies for the future.

# 📝 List of Sold Product 📝

> We provided a list of our most popular products, ordered by their total order quantities. This insight helps us identify our top-performing items, enabling us to strategize more effectively in terms of marketing and inventory management

```python
# @title  *(just in case if you want to know)* List of All Products { form-width: "8px", display-mode: "form" }
variable_name = ""
variable_name = ""
sold_products = sales['Product'].unique()

# Determine the number of products in the list
num_products = len(sold_products)

# Calculate the number of products per column (round up to ensure even distribution)
products_per_column = (num_products + 1) // 3

# Split the list of products into two columns
column1 = sold_products[:products_per_column]
column2 = sold_products[products_per_column:]
column3 = sold_products[2*products_per_column:]

# Combine the columns into a single list
sold_products_columns = list(zip(column1, column2, column3))

# Display the products in three columns
for col1, col2, col3 in sold_products_columns:
    print(f'{col1.ljust(40)} {col2.ljust(40)} {col3.ljust(40)}')

# If there are any remaining products, print them
if len(sold_products) % 3 != 0:
    for product in sold_products[3*products_per_column:]:
        print(product)
```

👇 **simple count of how many times each product appears** 👇

```python
sales['Product'].value_counts()
```

👇 **List Sold Products by Total Order Quantity, providing a ranked list of products from highest to lowest quantity ordered** 👇

```python
# Group the DataFrame by 'Product' and calculate the total order quantity for each product
product_quantity = sales.groupby('Product')['Order_Quantity'].sum().reset_index()

# Sort the products by total order quantity in descending order
sorted_products = product_quantity.sort_values(by='Order_Quantity', ascending=False)

# Display the sorted list of products
for index, row in sorted_products.iterrows():
    print(f'{row["Product"]}: {row["Order_Quantity"]}')
```

> 👓 In order to identify our top-selling products, we compiled a **list of sold products** that have consistently performed well. This information can guide us in maximizing our product range and marketing efforts.

# 💲 Correlation between Profit and Country ⛳

> 🔽 Lastly, we examined the potential correlation between profit and countries in which our stores located using **Boxplot**

> Boxplot particularly useful for identifying the central tendency and spread of data, as well as for detecting outliers🔽

```python
bp = sales.plot(kind='scatter', x='Country', y='Profit', figsize=(7,6))

plt.title('Correlation Between Profit and Country', fontsize=16)
plt.xticks(rotation=45)
plt.show(bp)
```

> 🔽  here is just another design using sns 🔽

```python
plt.figure(figsize=(7, 6))

sns.boxplot(x='Country', y='Profit', data=sales)

plt.title('Correlation Between Profit and Country', fontsize=16)
plt.xlabel('Country')
plt.ylabel('Profit')

# Rotate country label for better visibility
plt.xticks(rotation=45)
plt.show()
```

 > The grouped box plot analysis shows that the **stability in profit across different countries**, we should investigate further to understand the specific factors impacting profitability. This could involve a deeper analysis of cost structures, local market conditions, and potential areas for cost optimization or revenue enhancement.
